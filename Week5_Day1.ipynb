```python
import pandas as pd

names = ["Jess", "Jordan", "Sandy", "Ted", "Barney", "Tyler", "Rebecca"]
ages = [25, 36, 21, 28, 32, 27, 34]
income = [1000,2000,1200,3000,4000,2800,3400]
gender = ["Female", "Male", "Female", "Male", "Male", "Male", "Female"]
people = {"names":names, "ages":ages, "gender":gender, "Income(USD)":income}
df = pd.DataFrame.from_dict(people)

def incomeLevel(income):
    if income > 2000:
        return "High"
    else:
        return "Low"
    
df["Income_Level"] = df["Income(USD)"].apply(incomeLevel)
df.head(7)
```

```python
import pandas as pd

names = ["Jess", "Jordan", "Sandy", "Ted", "Barney", "Tyler", "Rebecca"]
ages = [25, 36, 21, 28, 32, 27, 34]
income = [1000,2000,1200,3000,4000,2800,3400]
gender = ["Female", "Male", "Female", "Male", "Male", "Male", "Female"]
people = {"names":names, "ages":ages, "gender":gender, "Income(USD)":income}
df = pd.DataFrame.from_dict(people)

def incomeLevel(income):
    if income > 2000:
        return "High"
    else:
        return "Low"
    
df["Income_Level"] = df["Income(USD)"].apply(incomeLevel)
df.groupby("Income_Level",as_index=True).count().head()
```

```python
df.groupby("Income_Level",as_index=True).mean().head()
```

```python
df.groupby("Income_Level",as_index=True).max().head() #Maxium
```

```python
df.groupby("Income_Level",as_index=True).min().tail() #Minium Income
```
