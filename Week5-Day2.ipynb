```python
from matplotlib import pyplot as plt
x, y = [1600, 1700, 1800, 1900, 2000] , [0.2, 0.5, 1.1, 2.2, 7.7]
plt.plot(x, y)
plt.title("World Population Over Time")
plt.xlabel("Year")
plt.ylabel("Population(billions)")
plt.show()
```

```python
x1, y1 = [1600, 1700, 1800, 1900, 2000] , [0.2, 0.5, 1.1, 2.2, 7.7]
x2, y2 = [1600, 1700, 1800, 1900, 2000] , [1, 1, 2, 3, 4]

plt.plot(x1, y1, "gx-", label="Actual")
plt.plot(x2, y2, "bo--", label="Fake")

plt.title("World Population Over Time")
plt.xlabel("Year")
plt.ylabel("Population(billions)")
plt.legend()
plt.show()
```

```python
x1, y1 = [1600, 1700, 1800, 1900, 2000] , [0.2, 0.5, 1.1, 2.2, 7.7]
x2, y2 = [1600, 1700, 1800, 1900, 2000] , [1, 1, 2, 3, 4]
x3, y3 = [1600, 1700, 1800, 1900, 2000] , [1, 4, 7, 8, 9]

plt.plot(x1, y1, "rx-", label="Actual")
plt.plot(x2, y2, "bo--", label="Fake")
plt.plot(x3, y3, "gx--", label="Guess")

plt.title("World Population Over Time")
plt.xlabel("Year")
plt.ylabel("Population(billions)")
plt.legend()
plt.show()
```

```python
from random import randint

height = [randint(58, 78) for x in range(20)]
weight = [randint(90, 250) for x in range(20)]

plt.scatter(weight, height)
plt.title("Height-Weight Distribution")
plt.xlabel("Weight (lbs)")
plt.ylabel("Height (inches)")
plt.show()
```

```python
import numpy as np
np.random.seed(5)
ages=[np. random.normal(loc = 40, scale=10) for x in range(1000)]
plt.hist(ages, bins=45)
plt.title("Ages per Population")
plt.xlabel("Age")
plt.ylabel("# of People")
plt.show()
```

```python
florida = [np.random.normal(loc=60, scale=15) for x in range(1000)]
california = [np.random.normal(loc=35, scale=5) for x in range(1000)]

plt.hist(florida, bins=45, color="r", alpha=0.4)

plt.hist(california, bins=45, color="b", alpha=0.4)

plt.show()

plt.hist(california, bins=45, color="g", alpha=1)
plt.show()
```

```python
from random import randint

height = [randint(58, 78) for x in range(20)]
weight = [randint(90, 250) for x in range(20)]
age = [randint(18, 65) for x in range(20)]

plt.scatter(weight, height, c=age)
plt.title("Height-Weight Distribution")
plt.xlabel("Weight (lbs)")
plt.ylabel("Height (inches)")
plt.colorbar(label="Age")
plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd

movieData = pd.read_csv("MovieData.csv")
movieData.columns = ["Film", "Genre", "CriticRatings", "AudienceRatings", "Budget", "Year"]
df = movieData.groupby("Year", as_index=False).sum().head()

x = df.Year
y = df.Budget

plt.plot(x, y)
plt.title("Budget Usage for Movies")
plt.xlabel("Year")
plt.ylabel("Total Budget(Millions)")
plt.show()
```

```python
x, y =[1600, 1700, 1800, 1900, 2000] , [0.2, 0.5, 1.1, 2.2, 7.7]
plt.plot(x, y, "bo-")
plt.title("World Population Over Time")
plt.xlabel("Year")
plt.ylabel("Population(billions)")
plt.savefig("population.jpg")
```

```python
df = movieData.groupby("Genre", as_index=False).count().head(20)

num_people = df.Budget
categories = df.Genre

plt.bar(categories, num_people)
plt.title("Movie Category")
plt.xlabel("Category")
plt.ylabel("Budget Usage (Millions)")
plt.show()
```

```python
df = movieData.groupby("Genre", as_index=False).count().head(100)
budget = df.Budget
category = df.Genre

plt.scatter(category, budget)
plt.title("Budget-Category Distribution")
plt.xlabel("Category")
plt.ylabel("Total Budget (millions)")
plt.show()
```

```python
df = movieData.groupby("Genre", as_index=False).count().head(100)
rating=df.AudienceRatings
year=df.Year
budget = df.Budget


plt.scatter(year, rating, c= budget)
plt.title("Movie Rating and Budget Over Time")
plt.xlabel("Year")
plt.ylabel("AudienceRatings")
plt.colorbar(label="Budget(Million)")
plt.grid(True)
plt.show()
```
